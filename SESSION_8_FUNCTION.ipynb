**Return statements**

- return statements are used to return the values

- every function will do some operations

- those operations will save in one variable

- if we want to use those values outside the function we can not use

- for that we are using **global** keyword

- we can also use those values outside the function using **return**

```python
def AVERAGE1(n1,n2,n3=20):
    summ=n1+n2+n3
    avg=summ/3
    print(avg)
    return(avg)
avg=AVERAGE1(10,20,30)    
```
